# 1463_1로 만들기

- dfs,bfs로 처리하기에는 시간 초과

---

```python
N = int(input())


# 1-1) 테이블 생성 
dp = [0] * (N+1)      

# 1-2) 초기값 처리
dp[1] = 0 


# 2) dp 테이블 채우기, 
# 3항 최솟값 연산, 2번 나눠서 처리 
for i in range(2, N+1):
    
    # 2-1) -1 default 처리
    dp[i] = dp[i-1] + 1
    
    # 2-2) 2의 배수라면, 
    if i % 2 == 0:      
        dp[i] = min( dp[i], dp[i//2] + 1 )
    
    # 2-3) 3의 배수라면, 
    if i % 3 == 0:      
        dp[i] = min( dp[i], dp[i//3] + 1 )

ans = dp[N]

print(ans)
```
